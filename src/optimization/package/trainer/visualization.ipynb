```python
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    0.3: 0.15927596356592075, 
    0.4: 0.15927596329899962, 
    0.5: 0.3919187358049589, 
    0.6: 0.4164527078098226, 
    0.7: 0.431470342142931, 
    0.8: 0.4329380264906126, 
    0.9: 0.38213521706352355
}

# Separate the keys and values for plotting
keys = list(data.keys())
values = list(data.values())

# Set style and context
sns.set_style("whitegrid")
sns.set_context("notebook", font_scale=1.25)

# Create the plot
plt.figure(figsize=(10, 6))
sns.barplot(x=keys, y=values, color=sns.color_palette()[0])

# Set labels and title
plt.xlabel('Threshold')
plt.ylabel('Dice')

# Display the plot
plt.tight_layout()
plt.show()
```
